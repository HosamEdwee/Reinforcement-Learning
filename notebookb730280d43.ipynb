# Mean-Variance Portfolio Optimization

Mean-Variance Portfolio Optimization is a process that aims to select the proportions of various assets to be included in a portfolio in such a way as to make the portfolio have desired properties, especially a good balance between expected return and risk.

The optimization problem can be formally expressed as:

$$
\begin{aligned}
& \underset{\mathbf{w}}{\text{minimize}}
& & \mathbf{w}^T \Sigma \mathbf{w} \\
& \text{subject to}
& & \mathbf{w}^T \mathbf{\mu} = \mu_p, \\
&&& \sum_{i=1}^{N} w_i = 1, \\
&&& w_i \geq 0, \; i = 1, \ldots, N.
\end{aligned}
$$

where:

- $\mathbf{w} = (w_1, \ldots, w_N)^T$ represents the weights of $N$ assets in the portfolio.
- $\Sigma$ is the covariance matrix of asset returns.
- $\mathbf{\mu} = (\mu_1, \ldots, \mu_N)^T$ is the vector of expected asset returns.
- $\mu_p$ is the desired expected return of the portfolio.
- $w_i$ is the weight of the $i^{th}$ asset in the portfolio.

The objective is to minimize the portfolio variance $\mathbf{w}^T \Sigma \mathbf{w}$, which is a measure of risk, while achieving a target expected return $\mu_p$ and ensuring that the sum of the weights equals 1 (i.e., all the capital is invested), with no short selling (weights are non-negative).
